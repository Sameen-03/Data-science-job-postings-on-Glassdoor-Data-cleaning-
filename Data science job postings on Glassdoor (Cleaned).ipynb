```python
#importing the necessary modules
import numpy as np #for linear algebra
import pandas as pd #for data processing
```

```python
#loading the data
df = pd.read_csv("C:/Users/ameen/Downloads/archive (2)/Uncleaned_DS_jobs.csv")
```

```python
#First lets remove the index column

df.drop('index', axis = 1, inplace = True)
```

```python
#modifying the salary column and creating a new column with average salaries

df['Salary Estimate'] = df['Salary Estimate'].map(lambda x: x.split('(')[0])
df['Salary Estimate'] = df['Salary Estimate'].str.replace("K","")

def a(x):
    x = x.split('-')
    min_salary = int(x[0].strip('$'))
    max_salary = int(x[1].strip('$'))
    average_salary = int((min_salary+max_salary)/2)
    return average_salary

df['Average salary'] = df['Salary Estimate'].apply(a)

#modifying the name of the column 'Salary Estimate'
df = df.rename(columns={'Salary Estimate': 'Salary Estimate (in thousands)'})
```

```python
# replacing the negative digits with zeroes in ratings column
df['Rating'] = df['Rating'].replace(-1.0,0)
```

```python
#Cleaning company name column
df['Company Name'] = df['Company Name'].map(lambda x: x.split('\n')[0])
```

```python
# modifying the 'size' column for efficiency of space
df['Size'] = df['Size'].str.replace('employees','')

#renaming the column name
df = df.rename(columns={'Size': 'Size (Number of employees)'})
```

```python
#replacing all the '-1' values in the data frame with 'na'
categorical_cols = ['Job Title', 'Salary Estimate (in thousands)', 'Job Description', 'Company Name', 'Location', 'Headquarters', 'Size (Number of employees)', 'Type of ownership', 'Industry', 'Sector', 'Revenue', 'Competitors']

df[categorical_cols] = df[categorical_cols].replace('-1', 'na')
df[categorical_cols] = df[categorical_cols].replace('Unknown', 'na')

df['Rating'] = df['Rating'].replace(-1., 0)
df['Founded'] = df['Founded'].replace(-1, 0)
```

```python
"""
The description column contains information about the skills which the interviewer is looking for
in their candidates / companies are looking for in their employees

We can create a bunch of columns for different skills contianing values ones and zeroes 
indicating the skill is required or not 
"""
df['python'] = df['Job Description'].map(lambda x: 1 if 'python' in x.lower() else 0)
df['hadoop'] = df['Job Description'].map(lambda x: 1 if 'hadoop' in x.lower() else 0)
df['excel'] = df['Job Description'].map(lambda x: 1 if 'excel' in x.lower() else 0)
df['aws'] = df['Job Description'].map(lambda x: 1 if 'aws' in x.lower() else 0)
df['tableau'] = df['Job Description'].map(lambda x: 1 if 'tableau' in x.lower() else 0)
df['powerBI'] = df['Job Description'].map(lambda x: 1 if 'powerbi' in x.lower() else 0)
```
